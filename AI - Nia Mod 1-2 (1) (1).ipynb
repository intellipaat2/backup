```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def model(epoch,alpha,activ):
  x=5
  y=10
  w=0.4
  b=0.12
  errorlist=[]
  y_predlist=[]
  wlist=[]
  blist=[]

  epoch=epoch
  alpha=alpha
  for i in range(epoch):
      #Feed Forward
      if activ=='None':
        y_pred= w*x + b
      else:
        y_pred=activ(w*x+b)
      y_predlist.append(y_pred)
      
      #Error
      error=(y- y_pred)**2
      errorlist.append(error)
      dedw=-2*x*(y-y_pred)
      dedb=-2*(y-y_pred)

      #Back Propagation
      w=w-alpha*dedw
      b=b-alpha*dedw
      wlist.append(w)
      blist.append(b)

  print(f'RMSE is {np.sqrt(np.mean(errorlist))} and last y_pred is {y_pred}')
  plt.figure(figsize=(10,8))
  plt.subplot(2,2,1)
  plt.plot(errorlist)
  plt.title("Error Graph")

  plt.subplot(2,2,2)
  plt.plot(y_predlist)
  plt.title('y_pred values')
```

```python
import pandas as pd
learning_rate = [1000,100,10,1,0.1,0.01,0.001,0.0001]
preds = []
rmse = []
for i in range(8):
  x = 5
  y = 10
  w = np.random.rand()
  b = np.random.rand()
  epoch = 30
  lr = learning_rate[i]
  for j in range(epoch):
    y_pred = w * x + b
    y_pred = round(y_pred, 2)
    error =round(y - y_pred, 2)
    rme = (error ** 2 ) ** 0.5
    dedw = -2 * x * error
    dedb = -2 * error
    w = w - lr * dedw
    b = b - lr * dedb
  preds.append(y_pred)
  rmse.append(rme)
df = pd.DataFrame({
    'x' : 8 * [x],
    'y' : 8 * [y],
    'y_pred' : preds,
    'rmse' : rmse,
    'Learning Rate' : learning_rate,
    'activation' : 8 * ['linear']
})
df
```

Single Perceptron Process

Input Layer: x = independent variable, y = dependent variable, w = weight, b = bias

Hidden Layer: w*x + b (This will be converted into activation function in case of non linearlity)

Output Layer: y_pred (predicted value for y)

a.

Feed Forward: Taking the variables from the input and performs the linear or non linear functions and gives out predicted output.

              w * x + b = y_pred
Error Calculation: Calculating the error by means of absolute difference between the predicted output and the auctual output.

          error = y - ypred
          square error = (error)2
          root square error = (square error)1/2
Back Propagation: Changing the weight and bias through differentiation with respect to error with the help of learning rate

         dedw = -2x(y - ypred)
         dedb = -2(y - ypred)
         w(revised) = w - alpha(learning rate) * dedw
         b(revised) = b - alpha(learning rate) * dedb
         y_pred = w(revised) * x + b(revised) 
After this, error calculation and back propagation is done repeatedly untill the error is reduced to the optimal level.

b.

Weight : It is the coefficient of the independent variable with respect to the dependent variable. It denotes the relation between independent and dependent variables.

Bias: It is y-intercept and represent the value of dependent variable when the independent variable is zero.

Learning Rate: It is a rate at which the model is learning the pattern in the problem. It directs the model towards the actual output. The number of iteration required to obtain the desired output depends on the learning rate

Q.No: 7. What are activation functions and why are they required in NN? a. Create python functions for sigmoid, tanh, relu & softmax

Activation functions are used to new dimensions in the data

Activation functions control the flowing numbers inside the neural network. Eg:tanh confines the values in neural networks within a range of (-1,1)

```python
#Sigmoid Function
def sig(a):
  a=(1/(1+((math.e)**(-a))))
  return a
#tanh Function
def tanh1(a):
  b=math.e**a
  c=math.e**-a
  d=(b-c)/(b+c)
  return d
#relu function
def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1
#Softmax function
def softmax1(a):
  blist=[]
  den=sum(math.e**i for i in (a))
  for i in a:
    d=math.e**i/den
    blist.append(d)
  return blist
```
