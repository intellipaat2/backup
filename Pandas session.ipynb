```python
def makeseries(a, b):
    return pd.Series(range(a, b+1))
```

```python
def makedf(lista , listb):
    return pd.DataFrame(listb, columns=lista)
```

```python
df3 = makedf(["One", "Two"], [["X", "Y"], ["A", "B"]])
df3
```

```python
df4 = makedf(["One", "Two"], [["x", "y"], ["a", "b"]])
df4
```

```python
def concatdfs(a,b):
    return pd.concat([a,b], ignore_index=True)
```

```python
new_df = concatdfs(df3, df4)
new_df
```
